La atenuacion en 3.5KHz coincide con es  de 0.5dB, por lo que cumple con la $\alpha_{max}$ en la banda de "Paso", y la atenuacion en 1KHz es de 37dB que cumple con la atenuacion mínima que se requiere en la banda de "Stop"

## Circuito Normalizado con $\Omega_\omega = 2\pi 3.5kHz$


## Simulacion 



```python
import numpy as np

import matplotlib.pyplot as plt

from scipy import signal as sig
from splane import pzmap, bodePlot, pretty_print_bicuad_omegayq

Wp = 2*np.pi*3.5e3
R1=1
R2=1
C1 = 99.64e-6
C2 = 41.069e-6
L1 = 29e-6
L2=70.47e-6


num = np.array(np.convolve([1,0,0],[1,0,0]))
den = np.array(np.convolve([1,R1/L1,1/(L1*C1)],[1,R2/L2,1/(L2*C2)]))


T = sig.TransferFunction(num , den)

plt.close('all')



bodePlot(T,[3,5],[1,-80], 'none', 'none', 'Filtro Pasa Altos', 1000 )
pzmap(T,True, 'Filtro Pasa Altos', 'none', 'none')



##TRANSFERENCIA NORMALIZADA
Omega_z=R1
Omega_w=Wp
R1_n=R1/Omega_z
R2_n = R2/Omega_z
C1_n = C1 * Omega_z * Omega_w
C2_n = C2 * Omega_z * Omega_w
L1_n = L1 * Omega_w / Omega_z
L2_n = L2 * Omega_w / Omega_z

num_n = np.array(np.convolve([1,0,0],[1,0,0]))
den_n = np.array(np.convolve([1,R1_n/L1_n,1/(L1_n*C1_n)],[1,R2_n/L2_n,1/(L2_n*C2_n)]))

T_n = sig.TransferFunction(num_n , den_n)



bodePlot(T_n,[-2,2],[1,-80], 'none', 'none', 'Filtro Pasa Bajos', 1000)
pzmap(T_n,True, 'Filtro Pasa Altos', 'none', 'none')
```
