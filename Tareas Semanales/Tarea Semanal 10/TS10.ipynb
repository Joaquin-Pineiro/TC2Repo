```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
##1)

num = np.array(np.convolve([1,0,3], [1,0,1]))
den = np.array(np.convolve([0,0,1], [1,0,2]))

w,Z = sig.freqs(num,den,1000)


plt.close('all')
fig,ax = plt.subplots(1,1,figsize=(10,5))
fig.suptitle('Impedancia Punto 1', fontsize=16)
ax.semilogx(w,20*np.log10(abs(Z)))
ax.set_xlabel(r'$\omega [rad/s]$')
ax.set_ylabel(r"$|Z(\omega)|$")
ax.grid(which='both', axis='both')
```

## Simulacion

```python
##2)

num = np.array(np.convolve([0,0,3], [1,0,7/3]))
den = np.array(np.convolve([1,0,2], [1,0,5]))


w,Y = sig.freqs(num,den,1000)


fig,ax = plt.subplots(1,1,figsize=(10,5))
fig.suptitle('Admitancia Punto 2', fontsize=16)
ax.semilogx(w,20*np.log10(abs(Y)))
ax.set_xlabel(r'$\omega [rad/s]$')
ax.set_ylabel(r"$|Y(\omega)|$")
ax.grid(which='both', axis='both')
```




### Se corrobora la existencia de dos polos en  $-\sqrt{2}$ y en  $-\sqrt{5}$ $\frac{rad}{s}$ y dos ceros en $-\sqrt{\frac{7}{3}}$ y en 0 $\frac{rad}{s}$
